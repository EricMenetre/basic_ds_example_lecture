```python
# We are now ready to write our function with proper type hints and documentation
def WH_data_for_given_year(WH_data_path: str, year: int) -> pd.DataFrame:
    ''' Read the World Happiness data at the provided path and extract the
        data corresponding to the specified year. Only columns 'Life Ladder',
        'Log GDP per capita', 'Social support', 'Healthy life expectancy at birth',
        'Freedom to make life choices', 'Generosity' and 'Perceptions of corruption'
        are kept.
    
        Parameters
        ----------
        WH_data_path
            The path to the World Happiness csv data
        year
            The year for which the data needs to be extracted

        Returns
        -------
        WH_df
            Data frame filtered by given year and columns of interest
    '''
    WH_df = pd.read_csv(WH_data_path)
    selec_rows_year = WH_df["Year"]==year
    selec_columns = WH_df.columns[2:9]
    return WH_df.loc[selec_rows_year, selec_columns]
```

We will now write a function taking the paths of the train and test data, together with a year, and predict the "Life Ladder" for that year. We will use this opportunity to rewrite the function to filter the dataframe by year to make it more generalizable.

```python
from typing import List
def WH_data_for_given_year(WH_data_path: str, ref_cols: List[str], year: int) -> pd.DataFrame:
    ''' Read the World Happiness data at the provided path and extract the
        data corresponding to the specified columns and year.
    
        Parameters
        ----------
        WH_data_path
            The path to the World Happiness csv data
        ref_cols
            A list of reference columns to keep
        year
            The year for which the data needs to be extracted

        Returns
        -------
        WH_df
            Data frame filtered by given year and columns of interest
    '''
    WH_df = pd.read_csv(WH_data_path)
    selec_rows_year = WH_df["Year"]==year
    return WH_df.loc[selec_rows_year, ref_cols]
```

```python
def predict_happiness_with_mlr(WH_train_data_path: str, WH_test_data_path: str,
                               year: int) -> pd.DataFrame:
    ''' Train a multiple linear regression model on a World Happiness training dataset
        for a given year, and make predictions on a test dataset for that year.
    
        Parameters
        ----------
        WH_train_data_path
            The path to the World Happiness csv training data
        WH_test_data_path
            The path to the World Happiness csv test data
        year
            The year for which the data needs to be extracted

        Returns
        -------
        WH_df
            Dataframe containing the test data with the predictions for the given year
            in a new column "Predicted Life Ladder"
    '''
    # Get the data
    train_cols = ['Life Ladder', 'Log GDP per capita', 'Social support', 
                  'Healthy life expectancy at birth', 'Freedom to make life choices',
                  'Generosity', 'Perceptions of corruption']
    test_cols = train_cols[1:] # all columns of train_cols except the first one
    WH_train_df = WH_data_for_given_year(WH_train_data_path, train_cols, year)
    WH_test_df = WH_data_for_given_year(WH_test_data_path, test_cols, year)
    # Clean them
    WH_train_clean_df = WH_train_df.dropna()
    WH_test_clean_df = WH_test_df.dropna()
    # Fit the model
    X_train = WH_train_clean_df.drop(columns="Life Ladder")
    y = WH_train_clean_df["Life Ladder"]
    lm = LinearRegression()
    lm.fit(X_train, y)
    # Predict scores and save them
    WH_test_clean_df["Predicted Life Ladder"] = lm.predict(WH_test_clean_df)
    path_withoutext, ext = os.path.splitext(WH_test_data_path)
    results_targetpath = path_withoutext + '_' + str(year) + '_predicted' + ext
    WH_test_clean_df.to_csv(results_targetpath, index=False)
    return results_targetpath
```

Often it is required to assess the performance of the model, i.e the "goodness of fit". For our linear model we could use R2 or the mean squared error for example. We will have a look at R2 now. 

In quite a few research paper, the R2 is reported using the same data for fitting the model and predicted the response. This can be misleading, as the true performance of interest in general is the estimation of the performance for new data or unseen data.

```python
# Let's compute R2 when we use our trained model to predict the same data we used to trained it 
# (this is an artifically best case performance and can be misleading, cf machine learning course next year)
# R2 is computed between the true y values and the predicted y values
from sklearn.metrics import r2_score
y_predicted_samedata = lm.predict(X_train)
r2_score(y, y_predicted_samedata)
```

In machine learning, this problem of over-estimation of the model performance (often due to over-fitting) is well known and a technique called cross-validation (CV) is used. In 5-fold CV your data is split in 5 equal parts called "folds". You set 1 fold aside, train your model on the remaining 4 folds, and then evaluate its performance on the unseed data in the fold you left aside. You can repeat this process 5 times (since there are 5 folds you can leave aside in turn). This way you not only have an estimation of the performance of your model on unseed data, but you can even obtain a confidence interval of this estimation since you have several performance measurements.

```python
# Let's have a look on how Kfold is creating indices to split our data in folds
from sklearn.model_selection import KFold
kf5 = KFold(n_splits=5)
for train_ix, test_ix in kf5.split(X):
    print("""Train set indices: {}\nTest set indices: {}""".format(train_ix, test_ix))
```

```python
# In addition of lookin at the performance at each of the 5 iterations
# we can look at the beta coefficient estimates for each of this iteration.
# Let's create a dataframe to save all these values
beta_R2_per_fold = pd.DataFrame(columns=list(X.columns)+['R2'], 
                                index=['fold1', 'fold2', 'fold3', 'fold4', 'fold5'])
beta_R2_per_fold
```

```python
# Now let's fill it by fitting the model for each of the train (4 folds) / test (1 fold) iterations
for i_fold, (train_ix, test_ix) in enumerate(kf5.split(X_train)):
    fold = 'fold{}'.format(i_fold+1)
    lm.fit(X_train.iloc[train_ix], y.iloc[train_ix])
    R2 =lm.score(X_train.iloc[test_ix], y.iloc[test_ix])
    beta_R2_per_fold.loc[fold] = list(lm.coef_) + [R2]
```

```python
n_cols = len(beta_R2_per_fold.columns)
plt.figure(figsize=(12,4))
for i_col, col in enumerate(beta_R2_per_fold.columns):
    plt.subplot(1, n_cols, i_col+1)
    sns.boxplot(y=col, data=beta_R2_per_fold)
plt.tight_layout()
```
